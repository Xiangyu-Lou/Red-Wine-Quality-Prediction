# Data Preprocessing

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Adjust Data Format

*Just for easy viewing*

```python
# path
file_path_data_raw = 'data/data_raw.csv'
file_path_data_format = 'data/data_format.csv'

# load data
column_names = [
    "fixed acidity", "volatile acidity", "citric acid", "residual sugar", "chlorides", "free sulfur dioxide", "total sulfur dioxide", "density", "pH", "sulphates", "alcohol", "quality"
]
data = pd.read_csv(file_path_data_raw, delimiter=';', names=column_names, header=0)

# Save the processed data
data.to_csv(file_path_data_format, index=False)

# Display
data.head()
```

# Check wheter the data has missing values

```python
# Check for missing values
missing_values = data.isnull().any()
missing_values
```

# Raw Data Visualization

## Boxplot

```python
fig, axs = plt.subplots(4, 3, figsize=(15, 20))

# Flatten the axes array
axs = axs.flatten()

# Iterate each feature
for i, col in enumerate(data.columns):
    # Box plot
    sns.boxplot(x=data[col], ax=axs[i])
    axs[i].set_title(f'Box Plot of {col}')

# Remove extra subplots
for i in range(len(data.columns), 12):
    fig.delaxes(axs[i])

plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap
fig2, ax2 = plt.subplots(figsize=(10, 7))
corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', ax=ax2)
ax2.set_title('Correlation Heatmap of Variables')
plt.show()
```

```python
fig, axs = plt.subplots(4, 3, figsize=(15, 20))

# Flatten the axes array
axs = axs.flatten()

# Iterate each feature
for i, col in enumerate(data.columns):
    # Histogram
    sns.histplot(data[col], kde=True, ax=axs[i])
    axs[i].set_title(f'Histogram of {col}')
    axs[i].set_xlabel(col)
    axs[i].set_ylabel('Frequency')

# Remove extra subplots
for i in range(len(data.columns), 12):
    fig.delaxes(axs[i])

plt.tight_layout()
plt.show()
```
